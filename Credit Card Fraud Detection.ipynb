**This plot helps analyze the distribution of each feature in the dataset and outliers**                                 
histplot is used because it effectively visualizes the distribution of numerical data. kde used for smooth curve for better understanding of the density.

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_temp = df.drop(['Time', 'Class', 'Amount'], axis=1)  # Dropping unnecessary columns
fig, ax = plt.subplots(ncols=4, nrows=7, figsize=(20, 50))  # Creating subplots

# Flattening the 2D array of axes for easier iteration
ax = ax.flatten()

for i, col in enumerate(df_temp.columns): #Used enumerate to iterate over both column names and subplot indices.
    sns.histplot(df_temp[col], kde=True, ax=ax[i])  # Plotting each column
    ax[i].set_title(f"Histogram of {col}")  # Setting title for each subplot

# Adjust layout after plotting
plt.tight_layout()
plt.show()


```

**Calculating the time Between fraud has been done**

```python
fraud_data = df[df['Class'] == 1]
fraud_start= fraud_data['Time'].min()
fraud_end= fraud_data['Time'].max()

#Convert to Hours and Minutes
start_hours = fraud_start // 3600
start_minutes = (fraud_start % 3600) // 60
end_hours = fraud_end // 3600
end_minutes = (fraud_end % 3600) // 60

print(f"Fraudulent transactions started at {start_hours} hours and {start_minutes} minutes.")
print(f"Fraudulent transactions ended at {end_hours} hours and {end_minutes} minutes.")
```

```python
time = pd.to_datetime(df['Time'], unit='s')

daily_transactions = df.groupby(time.dt.date).size()

transaction_counts =time.dt.date.value_counts().sort_index()
print(transaction_counts)

plt.figure(figsize=(12, 6))
daily_transactions.plot(kind='line', title="Daily Transactions Over Time", xlabel="Date", ylabel="Transaction Count")
```

**checking fraud activity occurance**

```python
df['hour'] = time.dt.hour
suspicious_hours1 = df[(df['hour'] >= 0) & (df['hour'] <= 6)]  # Transactions made between 0 AM and 6 AM
print(suspicious_hours1)
suspicious_hours2 = df[(df['hour'] >= 7) & (df['hour'] <= 12)]  # Transactions made between 7 AM and 12 AM
print(suspicious_hours2)
suspicious_hours3 = df[(df['hour'] >= 13) & (df['hour'] <= 18)]  # Transactions made between 13 AM and 18 AM
print(suspicious_hours3)
suspicious_hours4 = df[(df['hour'] >= 19) & (df['hour'] <= 24)]  # Transactions made between 19 AM and 24 AM
print(suspicious_hours4)
```
